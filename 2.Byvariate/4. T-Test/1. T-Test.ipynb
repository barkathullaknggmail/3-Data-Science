```python
dataset=dataset.dropna()
dataset
```

# Independent Paired T-Test
# Different group(male, female) but same condition(salary)

# Similarity between Male and Female Salary

```python
dataset[dataset['gender']=='M']['salary']
```

```python
dataset[dataset['gender']=='F']['salary']
```

```python
male=dataset[dataset['gender']=='M']['salary']
female=dataset[dataset['gender']=='F']['salary']
print("Male Salary",male)
print("FeMale Salary",female)
```

```python
from scipy.stats import ttest_ind
ttest_ind(male,female)
```

# p <0.05 so Reject Null Hypothesis and accept alternate hypothesis

# Interprete : There are  significant different between Male and Female Salary

# Similarity between Cenral and others board Salary

```python
dataset[dataset['hsc_b']== 'Others'	]['salary']
```

```python
dataset[dataset['hsc_b']== 'Central']['salary']
```

```python
cen=dataset[dataset['hsc_b']== 'Central']['salary']
ncen=dataset[dataset['hsc_b']== 'Others']['salary']
```

```python
from scipy.stats import ttest_ind
ttest_ind(cen,ncen)
```

# p>0.05 so Accept Null Hypothesis

# Intereprete : There are no significant diferent between Central and other board student salary package

# Dependent Paired T-Test
# Group(same -categorical data) but different condition(hsc_p,ssc_p Numerical data)

```python
male1=dataset[dataset['gender']=='M']['ssc_p']
male2=dataset[dataset['gender']=='M']['hsc_p']
ttest_ind(male1,male2)
```

# p >0.05 so accept Null hypothesis

# INTEREPRETE : THERE ARE NO SIGNIFICANT DIFFERENT BETWEEN 10TH AND PLUS TWO MARKE OF MALE STUDENT
