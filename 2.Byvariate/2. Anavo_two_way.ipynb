# Analysis of variance


# One way Clssification

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model=ols('salary~ssc_p', data=dataset).fit()
anova_result = sm.stats.anova_lm(model, type=2)
anova_result
```

```python
# Interpret the results
alpha = 0.05  # Significance level
if anova_result.iloc[0,4] < alpha:
    print("\nReject null hypothesis for SSLC Pass mark.")
else:
    print("\nAccept null hypothesis for SSLC Pass mark.")
```

# Two way Clssification

```python
model=ols('salary~ssc_p+hsc_p', data=dataset).fit()
anova_result = sm.stats.anova_lm(model, type=2)
anova_result
```

```python
if anova_result.iloc[0,4] < alpha:
    print("\nReject null hypothesis for SSLC Pass mark.")
else:
    print("\nAccept null hypothesis for SSLC Pass mark.")
if anova_result.iloc[1,4] < alpha:
    print("\nReject null hypothesis for HSC Pass mark.")
else:
    print("\nAccept null hypothesis for HSC Pass mark.")
```

```python
model=ols('salary~ssc_p+hsc_p+ssc_p*hsc_p', data=dataset).fit()
anova_result = sm.stats.anova_lm(model, type=2)
anova_result
```

```python
if anova_result.iloc[0,4] < alpha:
    print("\nReject null hypothesis for SSLC Pass mark.")
else:
    print("\nAccept null hypothesis for SSLC Pass mark.")

if anova_result.iloc[1,4] < alpha:
    print("Reject null hypothesis for HSC Pass mark.")
else:
    print("Accept null hypothesis for HSC Pass mark.")

if anova_result.iloc[2,4] < alpha:
    print("Reject null hypothesis for Interaction between SSLC and HSC.")
else:
    print("Accept null hypothesis for Interaction between SSLC and HSC.")
```

```python
model=ols('salary~ssc_p+hsc_p+degree_p+ssc_p*hsc_p*degree_p', data=dataset).fit()
anova_result = sm.stats.anova_lm(model, type=2)
anova_result
```

```python
# Interpret the results
alpha = 0.05  # Significance level
if anova_result.iloc[0,4] < alpha:
    print("\nReject null hypothesis for SSLC Pass mark.")
else:
    print("\nAccept null hypothesis for SSLC Pass mark.")

if anova_result.iloc[1,4] < alpha:
    print("Reject null hypothesis for HSC Pass mark.")
else:
    print("Accept null hypothesis for HSC Pass mark.")

if anova_result.iloc[2,4] < alpha:
    print("Reject null hypothesis for Degree Pass mark.")
else:
    print("Accept null hypothesis for Degree Pass mark.")

if anova_result.iloc[3,4] < alpha:
    print("Reject null hypothesis for Interaction between SSLC and HSC.")
else:
    print("Accept null hypothesis for Interaction between SSLC and HSC.")
if anova_result.iloc[4,4] < alpha:
    print("Reject null hypothesis for Interaction between SSLC and Degree.")
else:
    print("Accept null hypothesis for Interaction between SSLC and Degree.")
if anova_result.iloc[5,4] < alpha:
    print("Reject null hypothesis for Interaction between HSC and DEgree.")
else:
    print("Accept null hypothesis for Interaction between HSC and Degree.")
if anova_result.iloc[6,4] < alpha:
    print("Reject null hypothesis for Interaction between SSLC, HSC and DEgree.")
else:
    print("Accept null hypothesis for Interaction between SSLC, HSC and Degree.")
```
