```python
df2=pd.read_csv("placement.csv")
df2
```

```python
def qualQuan(dataset):
    quan=[]
    qual=[]
    for cn in dataset.columns:
        if(dataset[cn].dtype == 'O'):
            qual.append(cn)
        else:
            quan.append(cn)
    return qual, quan
```

```python
qual,quan=qualQuan(df2)
print(qual)
print(quan)
del quan[0]
print("After Delete Slno : ",quan)
```

```python
df3 = df2[quan]
df3
```

```python
df3.dtypes
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(df3)
df4=imp.transform(df3)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Creating an imputer object with mean strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Sample data with missing values
data = [[12, np.nan, 34], [10, 32, np.nan], [np.nan, 11, 20]]

print("Original Data:\n", data)

# Fitting the imputer on the data
imputer = imputer.fit(data)

# Transforming the data to replace missing values
data = imputer.transform(data)

print("Imputed Data:\n", data)
```
